```python
from time import time
from sys import setrecursionlimit
class PrintNumRec:
    setrecursionlimit(1000000000)
    def rec(self, x, y):
        if x == y:
            return
        else:
            print(x)
            return self.rec(x + 1, y)

    @staticmethod
    def init(n):
        start = time()

        y = 10 ** n
        x = y//10

        self = PrintNumRec()
        PrintNumRec.rec(self, x, y)
        print(f"For number {n}\nRecursive took: {time() - start} seconds")


PrintNumRec.init(10)
```
