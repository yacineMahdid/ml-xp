```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.random((800,3)))
df['date'] = pd.date_range('2000-1-1', periods=800, freq='D')
start_mask = ((df['date'].dt.month == 6) & (df['date'].dt.day >= 15)) | (df['date'].dt.month > 6)

end_mask = ((df['date'].dt.month == 8) & (df['date'].dt.day <= 15)) | (df['date'].dt.month < 8)

mask = start_mask & end_mask
print(df.loc[mask])

```
